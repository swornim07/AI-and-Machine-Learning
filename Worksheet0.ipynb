##1. Flatten a Nested List:
Write a Python function that takes a nested list and flattens it into a single list, where all the elements
are in a single dimension.
• Requirements:– Define a function flatten(lst) that takes a nested list lst and returns a flattened version
of the list.– Example: For the input [[1, 2], [3, 4], [5]], the output should be [1, 2, 3, 4, 5].

```python
def flatten(lst):
    """
    Flattens a nested list into a single list.

    Parameters:
        lst (list): A nested list containing sublists.

    Returns:
        list: A flattened list containing all elements.
    """
    flat_list = []

    for sublist in lst:
        for item in sublist:
            flat_list.append(item)

    return flat_list


# Example
numbers = [[1, 2], [3, 4], [5]]
result = flatten(numbers)
print(result)
```

Write a Python function that calculates the sum of all the numbers in a nested list (regardless of depth).

nested(lst) that takes a nested list lst and returns the sum of all

modified list.– Example: Fortheinput lst = [[1, 2], [3, 2], [4, 5]]andelem = 2,theoutputshould

```python
def remove_element(lst, elem):
    """
    Removes all occurrences of a specific element from a nested list.

    Parameters:
        lst (list): A nested list.
        elem: The element to remove.

    Returns:
        list: The modified nested list with the element removed.
    """
    result = []

    for item in lst:
        if isinstance(item, list):
            result.append(remove_element(item, elem))
        else:
            if item != elem:
                result.append(item)

    return result


# Example
lst = [[1, 2], [3, 2], [4, 5]]
elem = 2

print(remove_element(lst, elem))
```

##5. Find the Maximum Element in a Nested List:
Write a Python function that finds the maximum element in a nested list (regardless of depth).
• Requirements:– Define a function find
element.
max(lst) that takes a nested list lst and returns the maximum– Example: For the input [[1, 2], [3, [4, 5]], 6], the output should be 6

```python
def find_max(lst):
    """
    Finds the maximum value in a nested list.

    Parameters:
        lst (list): A nested list containing numbers.

    Returns:
        int/float: The maximum value found in the list.
    """
    maximum = None

    for item in lst:
        if isinstance(item, list):
            value = find_max(item)
        else:
            value = item

        if maximum is None or value > maximum:
            maximum = value

    return maximum


# Example
numbers = [[1, 2], [3, [4, 5]], 6]
result = find_max(numbers)
print(result)
```

##6. Count Occurrences of an Element in a Nested List:
Write a Python function that counts how many times a specific element appears in a nested list.
• Requirements:– Define a function count
occurrences(lst, elem) that counts the occurrences of elem in
the nested list lst.– Example: Fortheinput lst = [[1, 2], [2, 3], [2, 4]]andelem = 2,theoutputshould
be 3.

```python
def count_occurrences(lst, elem):
    """
    Counts how many times an element appears in a nested list.

    Parameters:
        lst (list): A nested list.
        elem: The element to count.

    Returns:
        int: The number of times the element appears.
    """
    count = 0

    for item in lst:
        if isinstance(item, list):
            count += count_occurrences(item, elem)
        else:
            if item == elem:
                count += 1

    return count


# Example
lst = [[1, 2], [2, 3], [2, 4]]
elem = 2

print(count_occurrences(lst, elem))
```

##7. Flatten a List of Lists of Lists:
Write a Python function that flattens a list of lists of lists into a single list, regardless of the depth.
• Requirements:– Define a function deep
flatten(lst) that takes a deeply nested list lst and returns a single
f
lattened list.– Example: For the input [[[1, 2], [3, 4]], [[5, 6], [7, 8]]], the output should be
[1, 2, 3, 4, 5, 6, 7, 8].

```python
def deep_flatten(lst):
    """
    Flattens a deeply nested list into a single list.

    Parameters:
        lst (list): A list that may contain other nested lists.

    Returns:
        list: A flattened list containing all elements.
    """
    result = []

    for item in lst:
        if isinstance(item, list):
            result.extend(deep_flatten(item))
        else:
            result.append(item)

    return result


# Example
numbers = [[[1, 2], [3, 4]], [[5, 6], [7, 8]]]
result = deep_flatten(numbers)
print(result)
```

of all the elements.

```python
def average_nested(lst):
    """
    Calculates the average of all elements in a nested list.

    Parameters:
        lst (list): A nested list containing numbers.

    Returns:
        float: The average of all numbers in the nested list.
    """

    flat_list = []

    def flatten(x):
        for item in x:
            if isinstance(item, list):
                flatten(item)
            else:
                flat_list.append(item)

    flatten(lst)

    return sum(flat_list) / len(flat_list)


# Example
numbers = [[1, 2], [3, 4], [5, 6]]
result = average_nested(numbers)
print(result)
```

```python
import numpy as np

# Given matrices
A = np.array([[3, 4],
              [7, 8]])

B = np.array([[5, 3],
              [2, 1]])

print("A:\n", A)
print("\nB:\n", B)

# ------------------------------------------------------------
# 1) Prove A * A^-1 = I
A_inv = np.linalg.inv(A)
I = np.eye(2)

print("\n1) A^-1:\n", A_inv)
print("\nA * A^-1:\n", A @ A_inv)
print("\nIs A * A^-1 close to I? ->", np.allclose(A @ A_inv, I))

# ------------------------------------------------------------
# 2) Prove AB != BA
AB = A @ B
BA = B @ A

print("\n2) AB:\n", AB)
print("\nBA:\n", BA)
print("\nIs AB equal to BA? ->", np.array_equal(AB, BA))

# ------------------------------------------------------------
# 3) Prove (AB)^T = B^T * A^T
left_side = (AB).T
right_side = B.T @ A.T

print("\n3) (AB)^T:\n", left_side)
print("\nB^T * A^T:\n", right_side)
print("\nAre they equal? ->", np.array_equal(left_side, right_side))

# ------------------------------------------------------------
# Solve the system of linear equations using inverse method
# 2x − 3y + z  = −1
# x  −  y + 2z = −3
# 3x +  y − z  =  9

# Matrix form: M * X = C
M = np.array([[2, -3,  1],
              [1, -1,  2],
              [3,  1, -1]])

C = np.array([-1, -3, 9])

print("\nCoefficient matrix M:\n", M)
print("\nConstant vector C:\n", C)

# Using inverse: X = inv(M) * C
M_inv = np.linalg.inv(M)
X = M_inv @ C

print("\nSolution [x, y, z]:\n", X)

# Optional check: M @ X should equal C
print("\nCheck (M @ X):\n", M @ X)
print("Is M @ X close to C? ->", np.allclose(M @ X, C))
```
